```python
from random import randint
import matplotlib.pyplot as plt
import time
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        return ((self.x - other.x) ** 2 + (self.y - other.y) ** 2) ** 0.5

    def __str__(self):
        return "({}, {})".format(self.x, self.y)
```

```python
def findClosest(n):
  points = []
  for i in range(n):
      points.append(Point(randint(0, n), randint(0, n)))

  min = n*2
  for i in range(n-1):
    for j in range(i+1, n) :
      distance = points[i].distance(points[j])
      if distance < min:
        min = distance
        pair = [points[i], points[j]]
  return points, min, pair
```

```python
points, min, pair = findClosest(60)
for point in points:
  if (point.x == pair[0].x and point.y == pair[0].y) or (point.x == pair[1].x and point.y == pair[1].y):
    plt.plot(point.x, point.y, 'gx')
  else:
    plt.plot(point.x, point.y, 'bo')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.show()
```

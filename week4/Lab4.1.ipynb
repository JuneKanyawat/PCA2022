```python
import time
import matplotlib.pyplot as plt
```

```python
def m1(n):
  rounds = 0
  sum = 0
  while rounds < n:
    sum += 1
    rounds +=1
  return sum

def m2(n):
  rounds = 0
  sum = 0
  while rounds < n:
    sum += 1
    rounds += 2
  return sum

def m3(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0
    while round2s < n:
      sum += 1
      round2s += 1
    round1s += 1
  return sum

def m4(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0
    while round2s < n:
      sum += 1
      round2s += 10
    round1s += 20
  return sum

def m5(n):
  round1s = 0
  sum = 0
  while round1s < n:
    sum += 1
    round1s += 1
  round2s = 0
  while round2s < n:
    sum += 1
    round2s += 1
  return sum

def m6(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0 
    while round2s < n*n:
      sum += 1
      round2s += 1
    round1s += 1
  return sum

def m7(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0
    while round2s < round1s:
      sum += 1
      round2s += 1
    round1s += 1
  return sum

def m8(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0
    while round2s < 100*round1s:
      sum += 1
      round2s += 1
    round1s += 1
  return sum

def m9(n):
  round1s = 0
  sum = 0
  while round1s < n:
    round2s = 0
    while round2s < n*n:
      round3s = 0
      while round3s < round2s:
        sum += 1
        round3s += 1
      round2s += 1
    round1s += 1
  return sum 

def m10(n):
  rounds = 1
  sum = 0
  while rounds < n:
    sum += 1
    rounds *= 2
  return sum

def m11(n):
  i = n 
  sum = 0
  while i > 0:
    sum += 1
    i /= 2
  return sum

def m12(n):
  rounds = 1
  sum = 0
  while rounds < n:
    sum += 1
    rounds *= 10
  return sum

def plotFunc(func, n):
  result = []
  for i in range(n):
    start = time.time()
    func(i)
    end = time.time()
    result.append(end - start)
  return result
```

```python
n = 8
plt.plot(plotFunc(m1, n), color='black')
plt.plot(plotFunc(m2, n), color='gray')
plt.plot(plotFunc(m3, n), color='brown')
plt.plot(plotFunc(m4, n), color='purple')
plt.plot(plotFunc(m5, n), color='indigo')
plt.plot(plotFunc(m6, n), color='blue')
plt.plot(plotFunc(m7, n), color='green')
plt.plot(plotFunc(m8, n), color='yellow')
plt.plot(plotFunc(m9, n), color='orange')
plt.plot(plotFunc(m10, n), color='red')
plt.plot(plotFunc(m11, n), color='pink')
plt.plot(plotFunc(m12, n), color='salmon')
plt.xlabel('n')
plt.ylabel('time')
plt.show()
```

```python
import matplotlib.pyplot as plt
from time import time
```

```python
def iterFib(n):
  a = 0
  b = 1
  if n == 1:
    return 1
  for i in range(n-1):
    c = a + b
    a = b
    b = c
  return c
```

```python
def fib(n):
    if(n <= 1):
        return n
    return fib(n-1) + fib(n-2)
```

```python
def plotFunc(func, n):
  result = []
  for i in range(n):
    start = time()
    func(n)
    end = time()
    result.append(end-start)
  return result
```

```python
plt.plot(plotFunc(iterFib, 30))
plt.ylim(0, 0.000008)
plt.show()
```

```python
plt.plot(plotFunc(fib, 30), color='purple')
plt.plot(plotFunc(iterFib, 30), color='blue')
plt.show()
```
